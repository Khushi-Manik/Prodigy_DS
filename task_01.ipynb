```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

uploaded = files.upload()

data = pd.read_csv('population.csv', skiprows=4)

print(data.head())
print(data.info())

latest_year = '2022'
population_data = data[['Country Name', latest_year]].dropna()

plt.figure(figsize=(12, 6))
sns.histplot(population_data[latest_year], kde=True, bins=30, color='blue')
plt.title(f'Distribution of Population ({latest_year})')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()

top_countries = population_data.nlargest(10, latest_year)

plt.figure(figsize=(14, 8))
sns.barplot(data=top_countries, x=latest_year, y='Country Name', palette='viridis')
plt.title(f'Top 10 Most Populated Countries ({latest_year})')
plt.xlabel('Population')
plt.ylabel('Country')
plt.show()
```
